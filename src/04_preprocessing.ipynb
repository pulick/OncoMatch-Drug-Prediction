```python
metadata_cols = ['SANGER_MODEL_ID', 'DRUG_ID', 'LN_IC50', 'DRUG_NAME', 'SMILES']
gene_cols = [c for c in df.columns if c not in metadata_cols]
```

```python
for col in gene_cols:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
nan_remaining = df.isna().sum().sum()
print(f"Missing values after imputation: {nan_remaining}")
df.to_parquet("C:/Users/abrup/Desktop/OncoMatch-Drug-Prediction/data/processed/AI_ready_matrix.csvAI_ready_matrix_FINAL.parquet", index=False)
```

```python
df=pd.read_csv("C:/Users/abrup/Desktop/OncoMatch-Drug-Prediction/data/processed/AI_ready_matrix.csv")
```

```python
print("-" * 30)
print(f"Matrix Shape: {df.shape}")
print(f"Number of Drugs: {df['DRUG_NAME'].nunique()}")
print(f"Number of Cell Lines: {df['SANGER_MODEL_ID'].nunique()}")
print(f"Number of Genes: {len(df.columns) - 5}")
nan_count = df.isna().sum().sum()
print("-" * 30)
print(f"Total Missing Values: {nan_count}")
if nan_count > 0:
    print("⚠️ Warning: You have missing values. You might need to impute them.")
print("-" * 30)
print("Target Variable (LN_IC50) Stats:")
print(df['LN_IC50'].describe())
```

```python
cols_list = df.columns[df.isna().any()].tolist()
```

```python
gene_cols = [c for c in df.columns if c not in ['SANGER_MODEL_ID', 'DRUG_NAME', 'SMILES', 'LN_IC50']]
variances = df[gene_cols].var().sort_values(ascending=False)

print("Top 10 Most Variable (Important) Genes:")
print(variances.head(10))

print("\nBottom 10 Least Variable (Useless) Genes:")
print(variances.tail(10))
```

```python


metadata_cols = ['SANGER_MODEL_ID', 'DRUG_NAME', 'SMILES', 'LN_IC50']
gene_cols = [c for c in df.columns if c not in metadata_cols]

# 3. Calculate Variance and pick the Top 2,000
print("Selecting top 2,000 most informative genes...")
variances = df[gene_cols].var().fillna(0) # Fill NaN variance with 0
top_genes = variances.sort_values(ascending=False).head(2000).index.tolist()

# 4. Create the Lean Matrix
lean_df = df[metadata_cols + top_genes]

# 5. Save for Training
lean_df.to_csv("C:/Users/abrup/Desktop/OncoMatch-Drug-Prediction/data/processed/AI_ready_matrix_LEAN.csv", index=False)

print(f"✅ Success! Matrix reduced from {df.shape[1]} to {lean_df.shape[1]} columns.")
print("The 'Noise' is gone. You are ready to build the model.")
```
