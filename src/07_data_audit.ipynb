```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:/Users/abrup/Desktop/OncoMatch-Drug-Prediction/data/processed/AI_ready_matrix.csv")
```

```python
df
```

```python
print("-" * 30)
print(f"Matrix Shape: {df.shape}")
print(f"Number of Drugs: {df['DRUG_NAME'].nunique()}")
print(f"Number of Cell Lines: {df['SANGER_MODEL_ID'].nunique()}")
print(f"Number of Genes: {len(df.columns) - 5}")
```

```python
nan_count = df.isna().sum().sum()
print("-" * 30)
print(f"Total Missing Values: {nan_count}")
if nan_count > 0:
    print("⚠️ Warning: You have missing values. You might need to impute them.")
```

```python
print("-" * 30)
print("Target Variable (LN_IC50) Stats:")
print(df['LN_IC50'].describe())
```

```python
random_gene = df.columns[100] 
print("-" * 30)
print(f"Example Gene ({random_gene}) Distribution:")
print(f"Min: {df[random_gene].min():.2f} | Max: {df[random_gene].max():.2f} | Mean: {df[random_gene].mean():.2f}")
```

```python
if df['SMILES'].iloc[0] is None:
    print("❌ ERROR: SMILES strings are missing!")
else:
    print("✅ SMILES strings are present.")
```

```python
cols_list = df.columns[df.isna().any()].tolist()
```

```python
cols_list
```

```python
gene_cols = [c for c in df.columns if c not in ['SANGER_MODEL_ID', 'DRUG_NAME', 'SMILES', 'LN_IC50']]
variances = df[gene_cols].var().sort_values(ascending=False)

print("Top 10 Most Variable (Important) Genes:")
print(variances.head(10))

print("\nBottom 10 Least Variable (Useless) Genes:")
print(variances.tail(10))
```
